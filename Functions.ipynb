```python
#User defined functions
#Without input and without return output
def mul():
  a=10
  b=20
  c = a*b
  print(c)

mul()
```

```python
#Funtion that takes input and does not return any output
def mul(x,y):
  c = x*y
  print(c)

mul(10,20)
```

```python
#Function that takes no input but returns output
def mul():
  a = 10
  b = 20
  c = a*b
  return c

res = mul()
print(res)
```

```python
#Function that takes input and returns output
def mul(x,y):
  c = x*y
  return c

res = mul(10,20)
print(res)
```

```python
#Positional arguments
#In the example we are performing power operation, here the argument are 2, 5 and parameters are a, b where a gets assigned as 2 and
#b gets assigned as 5 based on position.
def power_of(a,b):
   c = a**b
   print(c)

power_of(2, 5)
power_of(5, 2)#to verify this lets try to interchange the positions and check whether the value of c changes or remains same.
power_of(2)#if we miss a single argument then error appears
```

```python
#Default arguments
def power_of(a, x, b = 0):
  c=a*b*x
  print(c)


power_of(2, 5)
power_of(2, 5, 3)
```

```python
#Keyword arguments
#In positional arguments we noticed that if order changes the output also changes, we can overcome this by using keyword argument.
def power_of(a, b):
  c = a**b
  print(c)

power_of(a=2, b=5)
power_of(b=5, a=2)
```

```python
#Variable length arguments - passing any number of argument to a function is called as variable length argument
def pizza_toppings(toppings):
  print(toppings)

pizza_toppings("cheese")
```

```python
def pizza_toppings(*toppings):
  print(toppings)

pizza_toppings("cheese","onion","olives","corn")

# we can now see that by attaching a star or asterisk infront of the parameter, the function is now ready to take different number of arguments.
#This is because toppings is now considered as tuple
```

```python
def pizza_toppings(*toppings):
  print(toppings)
  print(type(toppings))

pizza_toppings("cheese","onion","olives","corn")

# * in front of the parameter makes it a tuple. But while calling for it we should just enter name which does not contain *.
```

```python
def pizza_toppings(*toppings, crust):
  print(toppings)
  print(crust)

pizza_toppings("cheese",crust = "thin")
```

```python
def pizza_toppings(*toppings, crust):
  print(toppings)
  print(crust)
pizza_toppings("cheese","thin")
# Here, we see that cheese and thin both are considr as toppings. so if we are passing any arguments before or after the variable length argument
# it is mandatory to use keyword arguments.
```

```python
#Variable keyword length argument

#In the previous exmp we saw how to pass diff number of arguments, but all of them were toppings.
#what if each argument we pass represents different data.
def collect_student_data(*data):
  print(data)

collect_student_data("Rohit", 20, 60.5,'M')
```

```python
# we certainly got output but we still can't recognize what is 28 , 60.5 , M.To resolve this we must give a key to each value entered as shown below.


def collect_student_data(*data):
  print(data)

collect_student_data(name = "rohit",age = 28, avg = 60.5, gender = 'M')
```

```python
#the error states that it doesn't recognise name and similarly age, avg, gender. To resolve this we must make another change.
#Let us see what that is, and how it resolves the issue
def collect_student_data(**data):
  print(data)

collect_student_data(name = "rohit",age = 28, avg = 60.5, gender = 'M')


#All we did was add another * to the parameter.
#And we can see in output that the keys and values are enclosed with {} which states ** infront of a parameter makes it a dictionary
```
