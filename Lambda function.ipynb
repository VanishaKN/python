```python
import logging

logging.debug("The debug message is displaying")
logging.info('The info msg is displaying')
logging.warning('The warning msg is displaying')
logging.error('The error msg is displaying')
logging.critical('The critical msg is displaying')
```

```python
#Let us see how to create one and how to call lambda funtion.
res = (lambda num,p : num**p)(2,5)
print(res)
```

```python
#What if we want to call the lambda function more than once? Is it possible ? Yes
fun = lambda num, deno : num/deno
res = fun(100,2)
print(res)
res1 = fun(10,2)
print(res1)
```

```python
#The normal function declaration using def along with filter() method
lst = [10,13, 18, 15, 20]
def fun(x):
    if(x%2 == 0):
      return True
    else:
      return False
evn_list = list(filter(fun,lst))
print(evn_list)
```

```python
#Using lambda function
lst = [10, 13, 18, 15, 20]
result = list((lambda x : (x%2 == 0),lst))
print(result)
```
